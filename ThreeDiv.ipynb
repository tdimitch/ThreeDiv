```python
import numpy as np


''' Values of the generating functions (completion of the curve) over exceptional divisors. '''

pol1 = [352,528,1056,880]
pol2 = [308,462,1155,1001]
pol3 = [336,560,1008,896]


''' Generate a matrix with exponent values, 
    the vanishing of the corresponding monomial over divisors and 
    minimal vanishing. '''

MonomioGen = [[0 for x in range(7)] for y in range(90)] 
counter = 0

for i in range(9):
    for j in range(5):
        for k in range(2):
            # exponents in x, y, z
            MonomioGen[counter][0] = i
            MonomioGen[counter][1] = j
            MonomioGen[counter][2] = k
            arr = [i,j,k,1]
            
            # jumping number condition in each divisor E_[1 2 3]
            MonomioGen[counter][3] = np.array(pol1) @ np.array(arr)
            MonomioGen[counter][4] = np.array(pol2) @ np.array(arr)
            MonomioGen[counter][5] = np.array(pol3) @ np.array(arr)
            
            # jumping number condition
            MonomioGen[counter][6] = min(MonomioGen[counter][3],MonomioGen[counter][4],MonomioGen[counter][5])
            counter += 1


''' Turn the matrix into an array for use of numpy functions 
    and sort the matrix by the minimal vanishing of the monomial. '''

MonomioGen = np.array(MonomioGen)
MonomioGen = MonomioGen[MonomioGen[:,6].argsort()]


''' Make a matrix (array) with elements of minimum vanishing < 3696 '''

# Number of monomials with minimum vanishing along divisors < 3696
number_dif_jn = sum(i < 3696 for i in MonomioGen[:,6])

# Keep those elements in JN
JN = MonomioGen[0:number_dif_jn,:]

# Number of different candidates, minimum vanishing
number_jn = len(set(JN[:,6]))


for i in range(44):
    # Compute where the minimum is attained
    divisors = np.where(JN[i][3:6] == np.amin(JN[i][3:]))
    
    # Print the set of jumping numbers < 1, together with the associated monomial and the divisors at which they are attained
    print("The monomial x^" + str(JN[i][0]) + "y^" + str(JN[i][1]) + 
         "z^" + str(JN[i][2]) + " generates the jumping number " + str(JN[i][6])+ 
         "/3696 at divisors E_", divisors[0]+1)    
    
    
print("The number of jumping numbers <1 in this list is: " + str(number_dif_jn))
print("The amount of different jumping numbers < 1 is: " + str(number_jn))
```
